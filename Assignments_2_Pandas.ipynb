#Counting the frequencies in a list using dictionary in Python

```python

def compute_dist(l1):

	# Creating an empty dictionary
	count = {}
	for item in l1:
		if (item in count):
			count[item] += 1
		else:
			count[item] = 1

	for key, value in count.items():
		print ("% d : % d"%(key, value))

if __name__ == "__main__":
	  l1 =[3, 4, 2, 1, 5, 1, 2, 4, 3, 4]

d1= compute_dist(l1)
```
